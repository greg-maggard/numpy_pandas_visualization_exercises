## 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x = range(100)

y1 = [math.sqrt(x) for x in x]
y2 = [x ** 3 for x in x]
y3 = [2 ** x for x in x]
y4 = [(1/(x+1)) for x in x]

plt.figure(figsize = (8, 6))

plt.subplot(221)
plt.plot(x, y1)
plt.title('$sqrt(x)$')

plt.subplot(222)
plt.plot(x, y2)
plt.title('${x}^3$')

plt.subplot(223)
plt.plot(x, y3)
plt.title('$2^{x}$')

plt.subplot(224)
plt.plot(x, y4)
plt.title('$1/(x+1)$')

plt.tight_layout()
```

## 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize = (8,6))
x = range(100)
plt.ylim(0, 100)
plt.xlim(0, 100)

y1 = [math.sqrt(x) for x in x]
y2 = [x ** 3 for x in x]
y3 = [2 ** x for x in x]
y4 = [(1/(x+1)) for x in x]

plt.plot(x, y1, c = 'Red', label = '$\sqrt{x}$')
plt.plot(x, y2, c = 'Orange', label = '${x}^3$')
plt.plot(x, y3, c = 'Green', label = '$2^{x}$')
plt.plot(x, y4, c = 'Blue', label = '$1/(x+1)$')

plt.title('Four Charts in One')
plt.legend()
```
